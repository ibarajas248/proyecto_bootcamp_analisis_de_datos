```python
import pandas as pd

# Estadísticas descriptivas básicas
print(df_partidos[['goles_local', 'goles_visitante']].describe())
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))

# Histograma de goles locales
plt.subplot(1, 2, 1)
plt.hist(df_partidos['goles_local'], bins=range(0, int(df_partidos['goles_local'].max()) + 2), edgecolor='black')
plt.title('Distribución de Goles Locales')
plt.xlabel('Goles Locales')
plt.ylabel('Frecuencia')

# Histograma de goles visitantes
plt.subplot(1, 2, 2)
plt.hist(df_partidos['goles_visitante'], bins=range(0, int(df_partidos['goles_visitante'].max()) + 2), edgecolor='black')
plt.title('Distribución de Goles Visitantes')
plt.xlabel('Goles Visitantes')
plt.ylabel('Frecuencia')

plt.tight_layout()
plt.show()

```

```python
df_partidos['fecha'] = pd.to_datetime(df_partidos['fecha'])  # Asegúrate de que 'fecha' esté en formato datetime

# Agrupa por fecha y suma los goles
goles_diarios = df_partidos.groupby('fecha')[['goles_local', 'goles_visitante']].sum()

# Visualiza las tendencias
plt.figure(figsize=(12, 6))
plt.plot(goles_diarios.index, goles_diarios['goles_local'], label='Goles Locales', marker='o')
plt.plot(goles_diarios.index, goles_diarios['goles_visitante'], label='Goles Visitantes', marker='o')
plt.title('Tendencia de Goles a lo Largo del Tiempo')
plt.xlabel('Fecha')
plt.ylabel('Número de Goles')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Agrupa por equipo visitante y local y calcula el promedio de goles
promedio_goles_local = df_partidos.groupby('local')['goles_local'].mean()
promedio_goles_visitante = df_partidos.groupby('visitante')['goles_visitante'].mean()

# Muestra los equipos con mejores y peores promedios de goles
print('Promedio de goles locales:')
print(promedio_goles_local.sort_values(ascending=False).head(10))

print('Promedio de goles visitantes:')
print(promedio_goles_visitante.sort_values(ascending=False).head(10))
```

```python
import pandas as pd

# Crear columnas para el resultado del partido
df_partidos['resultado_local'] = df_partidos.apply(
    lambda row: 'ganado' if row['goles_local'] > row['goles_visitante']
    else ('empatado' if row['goles_local'] == row['goles_visitante']
    else 'perdido'), axis=1
)

df_partidos['resultado_visitante'] = df_partidos.apply(
    lambda row: 'ganado' if row['goles_visitante'] > row['goles_local']
    else ('empatado' if row['goles_visitante'] == row['goles_local']
    else 'perdido'), axis=1
)
```

```python
# Contar resultados para equipos locales
resultados_local = df_partidos.groupby('local')['resultado_local'].value_counts().unstack(fill_value=0)
resultados_local = resultados_local.rename(columns={'ganado': 'ganados', 'empatado': 'empatados', 'perdido': 'perdidos'})

# Contar resultados para equipos visitantes
resultados_visitante = df_partidos.groupby('visitante')['resultado_visitante'].value_counts().unstack(fill_value=0)
resultados_visitante = resultados_visitante.rename(columns={'ganado': 'ganados', 'empatado': 'empatados', 'perdido': 'perdidos'})

# Sumar resultados de locales y visitantes
resultados = resultados_local.add(resultados_visitante, fill_value=0)
resultados = resultados.fillna(0)  # Asegurarse de que no haya valores nulos
```

```python
import matplotlib.pyplot as plt

# Crear un gráfico de barras horizontales
resultados.plot(kind='barh', stacked=True, figsize=(10, 8))

# Personalizar el gráfico
plt.title('Partidos Ganados, Empatados y Perdidos por Equipo')
plt.xlabel('Número de Partidos')
plt.ylabel('Equipo')
plt.legend(title='Resultado')
plt.grid(True)

# Mostrar el gráfico
plt.show()
```
